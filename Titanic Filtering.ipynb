```python
#replace null values in the 'Age' column with the mean age of passengers
df.loc[df['Age'].isnull(),'Age'] = df["Age"].mean()
```

```python
#replace null values in the 'Embarked' column with the most frequent value
df.loc[df['Embarked'].isnull(),'Embarked'] = df["Embarked"].mode()
```

```python
#How can you drop rows where all values are null in the Titanic dataset?
df.dropna()
```

```python
#Using .loc[], how can you filter rows for passengers whose 'Age' is greater than 30?
df.loc[df['Age']>30]
```

```python
#6. What command would you use with .iloc[] to select the first 10 rows and first 3 columns?
df.iloc[:10,:3]
```

```python
#. How can you filter the dataset to show only female passengers using .loc[]?
df.loc[df['Gender']=='female']
```

```python
#Using .loc[], how can you modify the 'Fare' of passengers with 'Pclass' 3 to 0 if the 'Fare' is null?
df.loc[(df['Pclass'] == 3) & (df['Fare'].isnull()),'Fare'] = 0
```

```python
df[(df['Pclass'] == 3) & (df['Fare'].isnull())]
```

```python
#How would you replace null values in multiple columns ('Age' and 'Fare') using their median values?
df.loc[df['Age'].isnull(),'Age'] = df['Age'].median()
df.loc[df['Fare'].isnull(),'Fare'] = df['Fare'].median()
```

```python
df[df['Age'].isnull()]
```

```python
df[df['Fare'].isnull()]
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
# Write a command to count the number of null values in each column of the Titanic dataset.
df.isnull().sum()
```

```python
#How can you use .iloc[] to retrieve data from rows 5 to 10 and columns 2 to 4?
df.iloc[4:10,1:4]
```
